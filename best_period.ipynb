```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python
df = pd.read_csv("40daylimit_allperiods.csv")

power_cols = [col for col in df.columns if re.match(r"LC\d+ Power", col)]

def find_best_lc(row):
    powers = row[power_cols]

    if powers.isna().all():
        return pd.Series([np.nan, np.nan, np.nan])

    best_power_col = powers.idxmax()

    lc_number = re.search(r"(LC\d+)", best_power_col).group(1)

    best_period = row[f"{lc_number} Period"]
    best_quarter = row[f"{lc_number} Quarter"]
    best_power = row[best_power_col]

    return pd.Series([best_power, best_period, best_quarter, lc_number])


df[["Best Power", "Best Period", "Best Quarter", "Best LC"]] = df.apply(find_best_lc, axis=1)

df.to_csv("best_periods_unchecked.csv", index=False)
```
