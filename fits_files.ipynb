Plot all light curves from a star given the folder path

```python
def plot_all_lightcurves(folder_path):
    fits_files = glob.glob(os.path.join(folder_path, "*.fits"))
    
    if len(fits_files) == 0:
        print("No FITS files found.")
        return
    
    i = 0

    for file in fits_files:
        print(f"Processing: {file}")
        lc = lk.read(file)
        lc = lc.select_flux("sap_flux")
        lc = lc.remove_nans().normalize()
        print(lc.meta['KEPLERID'])

        lc.plot()
        plt.show()
        i = i+1
        print(f"LC{i}")

folder = r"C:\Users\avery\Downloads\kepler_data\target_298.2520_40.4961\mastDownload\Kepler\kplr005301750_lc_Q011111111111111111"
plot_all_lightcurves(folder)
```
